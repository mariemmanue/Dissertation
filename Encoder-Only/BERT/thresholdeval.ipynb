```python
import pandas as pd
import numpy as np

# Path to the prediction TSV produced by eval.py
pred_path = "./data/results/answerdotai_ModernBERT-large_CGEdit_AAE_no-wandb-20260130-133308_CGEdit_AAE_FullTest_Final_preds.tsv"

df = pd.read_csv(pred_path, sep="\t")

# First column is the sentence, rest are feature probs
text_col = "sentence"  # adjust if your header name is different
head_list = list(df.columns[1:])  # all feature names
len(df), head_list[:5]
```

```python
all_texts = df[text_col].astype(str).tolist()
all_probs = df[head_list].to_numpy(dtype=float)  # shape (N, T)

all_probs.shape, all_texts[0]
```

```python
num_tasks = all_probs.shape[1]

for t in range(num_tasks):
    scores = all_probs[:, t]
    thr = np.percentile(scores, 90)  # top 10% as "positives"
    print(f"{head_list[t]}: 90th percentile threshold â‰ˆ {thr:.3f}")
```

```python
for t, feat in enumerate(head_list):
    scores = all_probs[:, t]
    sorted_idx = np.argsort(scores)  # ascending

    print(f"\n=== {feat}: lowest scores ===")
    for idx in sorted_idx[:3]:
        print(f"{scores[idx]:.4f} :: {all_texts[idx]}")

    print(f"\n=== {feat}: highest scores ===")
    for idx in sorted_idx[-3:][::-1]:
        print(f"{scores[idx]:.4f} :: {all_texts[idx]}")
```
